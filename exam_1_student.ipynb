## Regression 

You will now have to run a regression with sales as a function of TV. Show the summary statistics and check if the assumptions of linear regression are met (hint: use package gvlma).

```r
reg1 <- lm(TV ~ sales, data = data)
summary(reg1)
```

```r
gvlma(reg1)
```

Hierboven heb ik de regressie bekeken tussen "Sales" en "TV". De statistieken zijn samengevat hierboven te zien. Er is WEL/NIET voldaan aan de aanname van lineaire regressie.
